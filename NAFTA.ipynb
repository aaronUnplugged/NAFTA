```python
import pandas as pd
from pandas import Series,DataFrame
import numpy as np
from IPython.display import display, HTML

# For Visualization
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
%matplotlib inline

# For time stamps
from datetime import datetime

# For division
from __future__ import division
```

```python
#Import data from Excel
df=pd.read_csv('NAFTA.csv')
```

```python
display(df)
```

```python
list(df)
```

```python
# Delete extraneous rows at the top and bottom of the data set
df=df.ix[6:29]
display(df)
```

```python
#Plot trade with Mexico
df.plot(x='Year', y=['M_Imports','M_Exports'],figsize=(10,4))
```

```python
#Change data type to float
df['C_Imports']=df['C_Imports'].astype(float)
df['Year']=df['Year'].astype(float)
```

```python
#Plot trade with Canadian
df.plot(x='Year', y=['C_Imports','C_Exports'],figsize=(10,4))
```

```python
#Plot trade balance with Mexico and Canada
df.plot(x='Year', y=['M_Balance','C_Balance'],figsize=(10,4))
```

```python
#Plot US exports to each country
df.plot(x='Year', y=['M_Exports','C_Exports'],figsize=(10,4))
```

```python
#Plot US imports from each country
df.plot(x='Year', y=['M_Imports','C_Imports'],figsize=(10,4))
```

```python
#Add columns of Total Exports and Total Imports
df['Total_Exports']=df['M_Exports'] + df['C_Exports']
df['Total_Imports']=df['M_Imports'] + df['C_Imports']
```

```python
#Add column of Total US trade balance
df['Total_Balance']=df['Total_Exports'] - df['Total_Imports']
```

```python
#Add columns of total trade with each country
df['Total_Canada']=df['C_Exports'] + df['C_Imports']
df['Total_Mexico']=df['M_Exports'] + df['M_Imports']
```

```python
#Plot total trade with each country, Fig 1
fig, ax = plt.subplots()
plt.hold(True)
df['Total_Canada'].plot(x='Year', ax=ax, kind='bar', color='red',figsize=(12,5))
df['Total_Mexico'].plot(x='Year',kind='bar',color='blue',ax=ax)
ax.set_xlabel('Year')
ax.set_xticklabels(df.Year, rotation=90)
ax.set_ylabel('Total Trade (US$ Billions)')
ax.legend(loc='best')
```

```python
#Plot Total Exports to NAFTA countries from US and Total Imports from NAFTA countries to the US, fig 2
fig, ax = plt.subplots()
plt.hold(True)
df['Total_Exports'].plot(x='Year', ax=ax, kind='line', color='red',figsize=(12,5))
df['Total_Imports'].plot(x='Year',kind='line',color='blue',ax=ax)
ax.set_xlabel('Year')
ax.set_xticklabels([1991,1996,2001,2006,2011], rotation=90)
ax.set_ylabel('Total Trade (US$ Billions)')
ax.legend(loc='best')
```

```python
#Plot total US trade balance with NAFTA countries, fig 3
df.plot(x='Year', y='Total_Balance',color='blue',figsize=(12,5))

plt.axvline(x=1994, linewidth=2, color='red')
```

```python
#Plot Unemployment rate in each country, fig 4                        
fig, ax = plt.subplots()
plt.hold(True)
df['US_Unemployment'].plot(x='Year', ax=ax, kind='line', color='black',figsize=(12,5))
df['C_Unemployment'].plot(x='Year',kind='line',color='blue',ax=ax)
df['M_Unemployment'].plot(x='Year',kind='line',color='green',ax=ax)
ax.set_xlabel('Year')
ax.set_xticklabels([1991,1995,2000,2005,2010], rotation=90)
ax.set_ylabel('Unemployment (%)')
ax.legend(loc='best')

plt.axvline(x=16, linewidth=2, color='red')
plt.axvline(x=22, linewidth=2, color='red')
```

```python
#Plotting GDP per capita, fig 5

fig, ax = plt.subplots()
plt.hold(True)
df['US_GDPpc'].plot(x='Year', ax=ax, kind='line', color='black',figsize=(12,5))
df['C_GDPpc'].plot(x='Year',kind='line',color='blue',ax=ax)
df['M_GDPpc'].plot(x='Year',kind='line',color='green',ax=ax)
ax.set_xlabel('Year')
ax.set_xticklabels([1991,1995,2000,2005,2010], rotation=90)
ax.set_ylabel('GDP per Capita ($US Billions)')
ax.legend(loc='best')

plt.axvline(x=9, linewidth=2, color='red')
```

```python
#Calculate compound annual growth rates
CAGR_US = ((df.US_GDPpc[29])/(df.US_GDPpc[9]))**(1/20)-1
CAGR_CN = ((df.C_GDPpc[29])/(df.C_GDPpc[9]))**(1/20)-1
CAGR_MX = ((df.M_GDPpc[29])/(df.M_GDPpc[9]))**(1/20)-1

print ('GDP per capita in the US had a CAGR of {:.2%} '.format(CAGR_US))
print ('GDP per capita in Canada had a CAGR of {:.2%} '.format(CAGR_CN))
print ('GDP per capita in Mexico had a CAGR of {:.2%} '.format(CAGR_MX))
```

```python
#Plotting Total Factor Productivity, US as baseline, fig 6
fig, ax = plt.subplots()
plt.hold(True)
df['US_TFP'].plot(x='Year', ax=ax, kind='line', color='black',figsize=(12,5))
df['C_TFP'].plot(x='Year',kind='line',color='blue',ax=ax)
df['M_TFP'].plot(x='Year',kind='line',color='green',ax=ax)
ax.set_xlabel('Year')
ax.set_xticklabels([1991,1995,2000,2005,2010], rotation=90)
ax.set_ylabel('Total Factor Productivity')
ax.legend(loc='best')
axes = plt.gca()
axes.set_ylim([0.5,1.1])

plt.axvline(x=9, linewidth=2, color='red')
```

```python
#Change in Total Factor Productivity
TFP_CN = ((df.C_TFP[6])-(df.C_TFP[29]))/(df.C_TFP[6])
TFP_MX = ((df.M_TFP[6])-(df.M_TFP[29]))/(df.M_TFP[6])

print ('Total Factor Productivity in Canada declined by {:.2%} '.format(TFP_CN))
print ('Total Factor Productivity in Mexico declined by {:.2%} '.format(TFP_MX))
```
